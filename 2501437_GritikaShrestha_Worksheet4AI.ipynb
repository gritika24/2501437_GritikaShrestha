1.Load the dataset and perform EDA

```python
import pandas as pd
import numpy as np
dataframe = pd.read_csv("/content/drive/MyDrive/Worksheet4_dataset/diabetes_.csv")
print(dataframe.shape)
print(dataframe.head())
print(dataframe.columns)
print(dataframe.dtypes)
print(dataframe.describe())
print(dataframe.isnull().sum())
```

2. Handling missing data

```python
missing_info = dataframe.isnull().sum() / len(dataframe) * 100
for col in dataframe.columns:
  if missing_info[col]>10:
    dataframe[col].fillna(dataframe[col].median())
  else:
    dataframe.dropna(subset=[col],inplace=True)
print("Data after processing:n", dataframe.head())
```

3.Feature Engineering

Seperating data into test-train split

```python
X = dataframe.drop(columns=["Outcome"]).values
y= dataframe["Outcome"].values
#test-train split
def train_test_split(X, y, test_size=0.3, random_seed=42):
 np.random.seed(random_seed)
 indices = np.arange(X.shape[0])
 np.random.shuffle(indices)
 test_split_size = int(len(X) * test_size)
 test_indices = indices[:test_split_size]
 train_indices = indices[test_split_size:]
 X_train, X_test = X[train_indices], X[test_indices]
 y_train, y_test = y[train_indices], y[test_indices]
 return X_train, X_test, y_train, y_test
# Perform the train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
# Output shapes to verify
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)

```

Implementing KNN

```python
def euclidean_distance(point1, point2):
  if point1.shape != point2.shape:
    raise ValueError("Points must have same dimensions to calculate Euclidean distance")
  distance = np.sqrt(np.sum(point1 - point2) ** 2)
  return distance

```

Function to predict class for single query

```python
def knn_predict_single(query, X_train, y_train, k=3):
  distances = [euclidean_distance(query, x) for x in X_train]
  sorted_indices = np.argsort(distances)[:k]
  nearest_indices=sorted_indices[:k]
  nearest_labels = y_train[nearest_indices]
  prediction = np.bincount(nearest_labels).argmax()
  return prediction
```

Function to predict classes for all test samples

```python
def knn_predict(X_test, X_train, y_train, k=3):
  predictions = [knn_predict_single(x, X_train, y_train, k) for x in X_test]
  return np.array(predictions)
```

Evaluate using accuracy

```python
def compute_accuracy(y_true, y_pred):
  correct_predictions = np.sum(y_true == y_pred)
  total_predictions = len(y_true)
  accuracy = (correct_predictions / total_predictions) * 100
  return accuracy
```

```python
predictions = knn_predict(X_test, X_train, y_train, k=3)
print("Predictions:", predictions)
unscaled_accuracy = compute_accuracy(y_test, predictions)
print("Accuracy:", unscaled_accuracy)
```

Problem -3 Experiment with different values of k

```python
import matplotlib.pyplot as plt
import time
def experiment_knn_k_values(X_train, y_train, X_test, y_test, k_values):
  accuracies = {}
  times={}
  for k in k_values:
    start_time = time.time()
    predictions = knn_predict(X_test, X_train, y_train, k=k)
    accuracy = compute_accuracy(y_test, predictions)
    end_time = time.time()
    execution_time = end_time - start_time
    accuracies[k] = accuracy
    times[k]=execution_time
    print(f"Accuracy for k={k} is {accuracy:.2f}%")
    print(f"Time for k = {k} is {execution_time:.3f}seconds")
  return accuracies, times
k_values = range(1,16)
print("For unscaled data")
accuracy_unscaled, time_unscaled= experiment_knn_k_values(X_train, y_train, X_test, y_test, k_values)
print("For scaled data")
accuracy_scaled, time_scaled= experiment_knn_k_values(Xscaled_train, y_train, Xscaled_test, y_test, k_values)
```

Visualizing the results

```python
# Plot for accuracy
plt.figure()
plt.plot(k_values, accuracy_unscaled.values(), label="Unscaled", color="blue")
plt.plot(k_values, accuracy_scaled.values(), label="Scaled", color="red")
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.title("k vs Accuracy")
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot for time
plt.figure()
plt.plot(k_values, time_unscaled.values(), label="Unscaled", color="blue")
plt.plot(k_values, time_scaled.values(), label="Scaled", color="red")
plt.xlabel("k")
plt.ylabel("Time in seconds")
plt.title("k vs Execution time")
plt.legend()
plt.grid()
plt.show()
```

There are alot of challenges while using KNN for large datasets data one of them is that when a new data point is considered, KNN has to calculate the distance to all the training points which is not only expensive but takes time and memory

Additionally, the major challenges of high dimensional data is that it sufferes from the curse of dimensionality.When the features increases, the distance between the points become insignificant which makes it harder to find the true nearest data and reduces accuracy.

Strategies that can be implemented to improve efficiency of KNN is -

One of the strategies that can be implemented is reduction in dimensionality which causes the features to be reduced while still keeping important data's that makes it faster and reliable.

Similarly we can also use data structures like KD-tress or Ball tress which organizes data so that the searches for the nearest neighbours become faster.
