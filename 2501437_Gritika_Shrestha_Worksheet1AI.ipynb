```python
# 1) Empty 2x2 array
import numpy as np
arr = np.empty((2,2))
print(arr)


# 2) All ones array 4x2
arr = np.ones((4,2))
print(arr)

# 3) Array filled with a value
arr = np.full((4,2), 8)
print(arr)

# 4) Zeros array same shape as a given array
arr = [[1,2], [3,4]]
arr2 = np.zeros_like(arr)
print(arr2)

# 5) Ones array same shape as a given array
arr = [[1,2], [3,4]]
arr2 = np.ones_like(arr)
print(arr2)

# 6) Convert list to numpy array
new_list = [1,2,3,4]
arr = np.array(new_list)
print(arr)
```

Problem - 2: Array Manipulation: Numerical Ranges and Array indexing:

```python
# 1) Array from 10 to 49
arr = np.arange(10, 50)
print(arr)

# 2) 3x3 matrix 0 to 8
arr=np.arange(0,9)
reshape_arr=arr.reshape(3,3)
print(reshape_arr)

# 3) Identity matrix 3x3
arr= np.eye(3)
print(arr)

# 4) Random array of size 30 + mean
import random as random
arr = np.random.random(30)
print(arr)
mean= np.mean(arr)
print(f"mean is {mean}")

# 5) 10x10 random array min, max
arr= np.random.random((10,10))
max= np.max(arr)
min=np.min(arr)
print(f"max is {max} and min is {min}")

# 6) Zero array size 10, replace 5th element with 1
arr=np.zeros(10)
print(arr)
arr[4]=1
print(arr)

# 7) Reverse an array
arr=[1, 2, 0, 0, 4, 0]
reverse = np.flip(arr)
print(reverse)

# 8) Border 1s, inside 0s (example 5x5)
n,m = 5, 5
arr = np.zeros((n,m))
arr[0, :] = 1
arr[-1, :] = 1
arr[:, 0] = 1
arr[:, -1] = 1
print(arr)


# 9) 8x8 checkerboard
checkboard=np.ones((8,8))
checkboard[1::2, ::2]=0
checkboard[::2, 1::2]=0
print(checkboard)
```

4.2 Experiment: How Fast is Numpy?

Experiment: Python Lists vs NumPy Speed

Matrix Multiplication Test (1000Ã—1000)

```python
import timeit
list1 = list(range(1000000))
list2 = list(range(1000000))
```

```python
from timeit import timeit
def add_python(Z1, Z2):
  return [z1 + z2 for z1, z2 in zip(Z1, Z2)]
def add_numpy(Z1, Z2):
  return np.add(Z1, Z2)
Z1 = random.sample(range(1000000), 10000)
Z2 = random.sample(range(1000000), 10000)
python_time = timeit("add_python(Z1,Z2)", globals=globals(), number=100)
numpy_time = timeit("add_numpy(Z1,Z2)", globals=globals(), number=100)
print(f"Pure Python addition: {python_time:.2f} s per loop")
print(f"Numpy addition: {numpy_time:.2f} s per loop")
```

```python
from timeit import timeit
def mul_python(Z1, Z2):
  return [z1 * z2 for z1, z2 in zip(Z1, Z2)]
def mul_numpy(Z1, Z2):
  return np.multiply(Z1, Z2)
Z1 = random.sample(range(1000000), 10000)
Z2 = random.sample(range(1000000), 10000)
python_time = timeit("mul_python(Z1,Z2)", globals=globals(), number=100)
numpy_time = timeit("mul_numpy(Z1,Z2)", globals=globals(), number=100)
print(f"Pure Python multiplication: {python_time:.2f} s per loop")
print(f"Numpy multiplication: {numpy_time:.2f} s per loop")
```

```python
from timeit import timeit
def dot_python(Z1, Z2):
  return sum(z1 * z2 for z1, z2 in zip(Z1, Z2))
def dot_numpy(Z1, Z2):
  return np.dot(Z1, Z2)
Z1 = random.sample(range(1000000), 10000)
Z2 = random.sample(range(1000000), 10000)
python_time = timeit("dot_python(Z1,Z2)", globals=globals(), number=100)
numpy_time = timeit("dot_numpy(Z1,Z2)", globals=globals(), number=100)
print(f"Pure Python dot product: {python_time:.2f} s per loop")
print(f"Numpy dot product: {numpy_time:.2f} s per loop")
```
